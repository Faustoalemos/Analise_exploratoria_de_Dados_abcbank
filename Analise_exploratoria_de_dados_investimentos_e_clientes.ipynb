##  4. Transformao e tratamento dos dados.

### 4.1 Transformando os formatos de texto (object ou string) para inteiros (int64 ou float)

```python
data['poupanca'] = data['poupanca'].str.replace(',' , '.').astype(float)
data['cdb'] = data['cdb'].str.replace(',' , '.').astype(float)
data['renda'] = data['renda'].str.replace(',' , '.').astype(float)
```

#### Visualizando o resultado final da transformao dos formatos das colunas.

```python
data.info()
```

### 4.2 Criando colunas adicionais com dados de investimento total e participao;
#### Dessa forma iremos entender quanto porcento da renda cada cliente utiliza em produtos de investimento no banco.

```python
data['total investido'] = data['poupanca'] + data['cdb']
data['participacao'] = data['total investido'] / data['renda']
data
```

### 4.3 Descrio de cada coluna de forma estatstica

```python
data.describe()
```

### 4.4 Criao de uma nova coluna identificando grupos de clientes
#### Dessa forma  possvel visualizar os clientes que investem abaixo da mdia o que nesse caso iremos chamar de Grupo E

```python
def grupos_clientes (row):
    if row['participacao']>=0.40:
        return 'Grupo A'
    elif row['participacao']>=0.35:
        return 'Grupo B'
    elif row['participacao']>=0.30:
        return 'Grupo C'
    elif row['participacao']>=0.20:
        return 'Grupo D'
    else:
        return 'Grupo E'

data['Grupo'] = data.apply (grupos_clientes, axis=1)
data
```

### 4.5 Entendendo os volumes de clientes por grupo.

```python
#Vamos analisar o volume por grupo que foi gerado na coluna Grupo, e j conseguimos entender que existem muitos clientes que se tornam
#possveis o contato para venda de produtos de investimentos.
data.Grupo.value_counts()
```

```python
data.info()
```

### 4.6 Analisando a base novamente e projetando os possveis retornos com investimento clientes grupo E
   #### 4.6.1 Entendendo a mdia de participao de investimento dos clientes.

```python
#Aps a anlise podemos visualizar e projetar os possveis ganhos se realmente cada cliente alcanar a mdia de investimento padro.
media = data['participacao'].mean()
media
```

   #### 4.6.2 Criando uma nova coluna projetando a participao dos clientes grupo E com 18% da sua renda e convertendo em valores.

```python
data['Projecao'] = np.where(
    data['Grupo'] == 'Grupo E',
    data['renda'] * 0.18,
    0)
```

```python
data
```

  #### 4.6.3 Efetuando a soma total dos clientes Grupo E em investimentos no Banco que somam o total de R$ 275.937.779,88

```python
data['Projecao'].sum()
```
